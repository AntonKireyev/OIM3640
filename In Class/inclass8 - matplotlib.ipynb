```python
from math import exp, sqrt
import matplotlib.pyplot as plt
from random import normalvariate as norm
```

```python
r = 0.0525
T = 1/252
sigma = 0.30
appl = 187
prices = list()
for day in range(63):
    prices.append(appl)
    appl *= exp((r - 0.5 * sigma ** 2)*T + sigma * sqrt(T) * norm(0,1))
```

```python
prices[0:10]
```

```python
plt.plot(prices, label = 'AAPL', ls = "-.")       # Create graph, include label, line style, etc.
plt.title("Possible Price Path for Appl", fontsize = 16)        # Title
plt.xlabel("Days")      # X label
plt.ylabel("Price")     # Y label
plt.grid(axis = "y", alpha = 0.6)    # Grid lines (nothing = both axis, you can also specify axis)
plt.legend(loc = 2);        # can set locations for the label, move it around. add semicolon to suppress text output (memory location)

```
